Downloaded Hourly Temperature Data for RIVERSIDE MUNICIPAL AIRPORT, CA US from https://www.ncei.noaa.gov/access/search/data-search/local-climatological-data?bbox=34.130,-117.526,33.832,-117.228&pageNum=1&startDate=2015-01-03T23:59:59&endDate=2024-01-03T00:00:00.
Data was downloadeble for each year, therefore, downloaded for years 2015-2024 and concatanated them and made a combined file for hourly temperature dataset for Riverside airport for 2015-2024.

```python
import pandas as pd
import glob

# Create a list of file names in order
file_names = [f'{i}.csv' for i in range(15, 25)]

# Create an empty list to store the DataFrames
all_dfs = []

# Loop through the file names, read each CSV, and append to the list
for file_name in file_names:
    df = pd.read_csv(file_name)
    all_dfs.append(df)

# Concatenate all DataFrames in the list
combined_df = pd.concat(all_dfs, ignore_index=True)

# Save the combined DataFrame to a new CSV file
combined_df.to_csv('combined.csv', index=False)
```

```python
 # Read the file path
 csv_path= "/content/combined.csv"
```

```python
# Read the file into a dataframe
df = pd.read_csv(csv_path)
```

```python
# Converting 'DATE' column to datetime objects and formatting
df['DATE'] = pd.to_datetime(df['DATE']) # Converting 'DATE' column to datetime
df['DATE'] = df['DATE'].dt.strftime('%Y-%m-%d %H:%M:%S') # Formatting the datetime objects
```

```python
df # Displaying the DataFrame
```

```python
df.index = pd.to_datetime(df['DATE'], format='%Y-%m-%d %H:%M:%S') # Setting 'DATE' column as index
df[:26] # Displaying the first 26 rows of the DataFrame
```

```python
# Checking data types of columns
print("Data types of columns:")  # Printing a message to indicate the data types output
print(df.dtypes)  # Displaying data types of columns
```

```python
# Converting 'HourlyWetBulbTemperature' to numeric and handling errors
df['HourlyWetBulbTemperature'] = pd.to_numeric(df['HourlyWetBulbTemperature'],errors = 'coerce') # Converting column to numeric
```

```python
# Displaying information about the DataFrame
df.info() # Displaying information about the DataFrame
```

```python
# Filling missing values in 'HourlyWetBulbTemperature' with the mean
df['HourlyWetBulbTemperature'].fillna(int(df['HourlyWetBulbTemperature'].mean()), inplace=True) # Filling missing values with the mean
```

```python
# Plotting 'HourlyWetBulbTemperature'
Temperature = df['HourlyWetBulbTemperature'] # Assigning the column to a variable
Temperature.plot() # Plotting the values
```

```python
#[[[E1],[E2],[E3],...,[E50]]] [E51]
#[[[E2],[E3],[E4],...,[E51]]] [E52]
#[[[E3],[E4],[E5],...,[E52]]] [E53]

# Function to convert DataFrame to X and y arrays for LSTM model
def df_to_X_y(df, window_size=30):
  df_as_np = df.to_numpy() # Converting DataFrame to NumPy array
  X = [] # Initializing empty list for X
  y = [] # Initializing empty list for y
  for i in range(len(df_as_np)-window_size):
    row = [[a] for a in df_as_np[i:i+30]] # Creating a row of data for X
    X.append(row) # Appending the row to X
    label = df_as_np[i+30] # Getting the label for y
    y.append(label) # Appending the label to y
  return np.array(X), np.array(y) # Returning X and y as NumPy arrays
```

```python
window_size = 30 # Setting the window size
X, y = df_to_X_y(Temperature, window_size) # Calling the function to create X and y
X.shape, y.shape # Displaying the shape of X and y
```

```python
X # Displaying the X array
```

```python
y # Displaying the y array
```

```python
# Splitting data into train, validation, and test sets
X_train, y_train = X[:75000], y[:75000] # Training data
X_val, y_val = X[75000:90000], y[75000:90000] # Validation data
X_test, y_test = X[90000:], y[90000:] # Test data
X_train.shape, y_train.shape, X_val.shape, y_val.shape, X_test.shape, y_test.shape # Displaying the shapes
```

```python
# Importing necessary libraries for LSTM model
from tensorflow import keras
from keras.models import Sequential
from keras.layers import *
from keras.callbacks import ModelCheckpoint
from keras.losses import MeanSquaredError
from keras.metrics import RootMeanSquaredError
from keras.optimizers import Adam

# Building the LSTM model

model1 = Sequential() # Creating a sequential model
model1.add(InputLayer((30,1))) # Adding an input layer
model1.add(LSTM(128)) # Adding an LSTM layer
model1.add(Dense(32, 'relu')) # Adding a dense layer with ReLU activation
model1.add(Dense(1, 'linear')) # Adding an output layer with linear activation

model1.summary() # Displaying the model summary
```

```python
# Setting up model checkpoint and compiling the model
cp = ModelCheckpoint('model1.model1.keras', save_best_only=True) # Creating a model checkpoint
model1.compile(loss = MeanSquaredError(), optimizer=Adam(learning_rate=0.00001), metrics=[RootMeanSquaredError()]) # Compiling the model
```

```python
# Training the model
model1.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=10, callbacks=[cp]) # Training the model
```

```python
# Loading the trained model
from keras.models import load_model
model1 = load_model('model1.model1.keras') # Loading the trained model
```

```python
train_predictions = model1.predict(X_train).flatten() # Making predictions
train_results = pd.DataFrame(data={'Train Predictions':train_predictions, 'Actuals' :y_train}) # Creating a DataFrame for results
train_results # Displaying the training results
```

```python
# Plotting training predictions vs actual values
# Actual plot is blue
# Prediction plot is orange
import matplotlib.pyplot as plt
plt.plot(train_results['Actuals'][:1000]) # Plotting actual values
plt.plot(train_results['Train Predictions'][:1000]) # Plotting predictions
```

```python
# Making predictions on the validation set
val_predictions = model1.predict(X_val).flatten() # Making predictions
val_results = pd.DataFrame(data={'Val Predictions':val_predictions, 'Actuals' :y_val}) # Creating a DataFrame for results
val_results # Displaying the validation results
```

```python
# Plotting validation predictions vs actual values
# Actual is blue
# Prediction plot is orange
plt.plot(val_results['Actuals'][:1000]) # Plotting actual values
plt.plot(val_results['Val Predictions'][:1000]) # Plotting predictions
```

```python
# Making predictions on the test set
test_predictions = model1.predict(X_test).flatten() # Making predictions
test_results = pd.DataFrame(data={'test Predictions':test_predictions, 'Actuals' :y_test}) # Creating a DataFrame for results
test_results # Displaying the test results
```

```python
# Plotting test predictions vs actual values
# Actual is blue
# Prediction plot is orange
plt.plot(test_results['Actuals'][:1000]) # Plotting actual values
plt.plot(test_results['test Predictions'][:1000]) # Plotting predictions
```
